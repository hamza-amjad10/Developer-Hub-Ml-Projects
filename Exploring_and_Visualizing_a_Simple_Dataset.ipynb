```python
plt.scatter(x=df['sepal_length'],y=df["sepal_width"])
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")
plt.show()
```

```python
plt.scatter(x=df['petal_length'],y=df["petal_width"])
plt.xlabel("petal_length")
plt.ylabel("petal_width")
plt.show()
```

```python
df["species"]=df["species"].map({"Iris-setosa":0,"Iris-versicolor":1,"Iris-virginica":2})
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].scatter(x=df['sepal_length'], y=df['sepal_width'], c=df['species'], cmap='Set1')
axes[0].set_xlabel('sepal_length')
axes[0].set_ylabel('sepal_width')
axes[0].set_title('Sepal Length vs Width')

axes[1].scatter(x=df['petal_length'], y=df['petal_width'], c=df['species'], cmap='Set1')
axes[1].set_xlabel('petal_length')
axes[1].set_ylabel('petal_width')
axes[1].set_title('Petal Length vs Width')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].hist(df["sepal_length"])   # sahi method aur column name
axes[0, 0].set_title("Sepal length")

axes[0, 1].hist(df["sepal_width"])   # sahi method aur column name
axes[0, 1].set_title("Sepal Width")

axes[1, 0].hist(df["petal_length"])   # sahi method aur column name
axes[1, 0].set_title("Petal Length")

axes[1, 1].hist(df["petal_width"])   # sahi method aur column name
axes[1, 1].set_title("Petal Width")

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Sepal Length
axes[0, 0].boxplot(df['sepal_length'])
axes[0, 0].set_title('Sepal Length')

# Sepal Width
axes[0, 1].boxplot(df['sepal_width'])
axes[0, 1].set_title('Sepal Width')

# Petal Length
axes[1, 0].boxplot(df['petal_length'])
axes[1, 0].set_title('Petal Length')

# Petal Width
axes[1, 1].boxplot(df['petal_width'])
axes[1, 1].set_title('Petal Width')

plt.show()
```
